```python
import nltk
from nltk.stem import wordnet, PorterStemmer, WordNetLemmatizer
from nltk import pos_tag, word_tokenize, sent_tokenize
import pandas as pd
import numpy as np
from sklearn.feature_extraction.text import CountVectorizer
```

```python
text = [
    'Hello my name is james, this is my python notebook',
    'this is my python notebook',
    'james trying to create a big dataset',
    'james of words to try differnt',
    'features of count vectorizer'
    ]
```

```python
count_vectorizer = CountVectorizer()
count_vectorizer_matrix = count_vectorizer.fit_transform(text)
count_vectorizer_array = count_vectorizer_matrix.toarray()
df = pd.DataFrame(count_vectorizer_array, columns=count_vectorizer.get_feature_names_out())
df
```

#### lowercase by default True , set it to false

```python
count_vectorizer = CountVectorizer(lowercase = False)
count_vectorizer_matrix = count_vectorizer.fit_transform(text)
count_vectorizer_array = count_vectorizer_matrix.toarray()
df = pd.DataFrame(count_vectorizer_array, columns=count_vectorizer.get_feature_names_out())
df
```

#### StopWords
- Cusome Stop Words

```python
count_vectorizer = CountVectorizer(lowercase = False, stop_words=['is', 'to', 'my'])
count_vectorizer_matrix = count_vectorizer.fit_transform(text)
count_vectorizer_array = count_vectorizer_matrix.toarray()
df = pd.DataFrame(count_vectorizer_array, columns=count_vectorizer.get_feature_names_out())
df
```

- Sklearn Stop Words List

```python
count_vectorizer = CountVectorizer(lowercase = False, stop_words='english')
count_vectorizer_matrix = count_vectorizer.fit_transform(text)
count_vectorizer_array = count_vectorizer_matrix.toarray()
df = pd.DataFrame(count_vectorizer_array, columns=count_vectorizer.get_feature_names_out())
df
```
