```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.svm import SVR
import os
%matplotlib inline
```

```python
X = np.sort(np.random.rand(40,1)*5,axis=0)
```

```python
y = np.sin(X).ravel()
```

```python
y[::5] += 3 * (0.5 - np.random.rand(8))
```

```python
clf = SVR(kernel='linear')
clf.fit(X,y)
y_preds = clf.predict(X)
plt.scatter(X,y, color='darkorange',
            label='data')
plt.plot(X, y_preds, color='cornflowerblue',
         label='prediction')
plt.legend()
plt.show()
```

```python
clf = SVR(kernel='poly')
clf.fit(X,y)
y_preds = clf.predict(X)
plt.scatter(X,y, color='darkorange',
            label='data')
plt.plot(X, y_preds, color='cornflowerblue',
         label='prediction')
plt.legend()
plt.show()
```

```python
clf = SVR(kernel='rbf')
clf.fit(X,y)
y_preds = clf.predict(X)
plt.scatter(X,y, color='darkorange',
            label='data')
plt.plot(X, y_preds, color='cornflowerblue',
         label='prediction')
plt.legend()
plt.show()
```

```python
clf = SVR(kernel='sigmoid')
clf.fit(X,y)
y_preds = clf.predict(X)
plt.scatter(X,y, color='darkorange',
            label='data')
plt.plot(X, y_preds, color='cornflowerblue',
         label='prediction')
plt.legend()
plt.show()
```

### Real Data Demo 00

```python
DATA_PATH = os.path.join(os.getcwd(),'data','Position_Salaries.csv')
df = pd.read_csv(DATA_PATH)
df.head()
```

```python
df.isna().sum()
```

```python
df.columns
```

```python
X = df.iloc[:,1:2].values.astype('f')
y = df.iloc[:,2:3].values.astype(np.float64)
```

```python
# Scale Featutes
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVR
from sklearn.metrics import accuracy_score
```

```python
scl = StandardScaler()
X = scl.fit_transform(X)
y = scl.fit_transform(y)

for k in ['linear', 'poly', 'rbf', 'sigmoid']:
    clf = SVR(kernel=k)
    clf.fit(X,y)
    #5 Predicting a new result
    y_preds = clf.predict(X)
    
    plt.scatter(X,y,label=f"Real Data {k}",color='orange')
    plt.plot(X,y_preds,label=f"Prediction Data {k}")
    plt.legend()
    plt.show()
```

### Real Demo With SVR 01

```python
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
```

```python
iris.feature_names
```

```python
X =iris.data[:,:2]
y =iris.target
```

```python
y
```

```python
from sklearn.svm import SVC
```

```python
x_min,x_max = X[:,0].min()-1,X[:,0].max()+1
y_min,y_max = X[:,1].min(),X[:,1].max()+1
h = (x_max / x_min)/100
for k in ['linear', 'poly', 'rbf', 'sigmoid']:
    
    svc = SVC(kernel=k)
    svc.fit(X,y)
    #Prepare Space to droaw boundries
    xx,yy = np.meshgrid(np.arange(x_min,x_max,h),np.arange(y_min,y_max,h))
    #contourse between classes
    Z = svc.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)
    
    plt.subplot(1,1,1)
    plt.contourf(xx,yy,Z,map=plt.cm.Paired, alpha=0.8)
    plt.scatter(X[:, 0], X[:, 1], c=y, cmap=plt.cm.Paired,label=k)

    plt.legend()
    plt.xlabel(f'Sepal length kernel {k} ')
    plt.ylabel(f'Sepal width kernel {k} ')
    plt.xlim(xx.min(), xx.max())
    plt.title(f'SVC with kernel {k} ')
    plt.show()
```
